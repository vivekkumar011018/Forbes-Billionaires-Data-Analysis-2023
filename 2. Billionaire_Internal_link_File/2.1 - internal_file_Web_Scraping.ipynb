# Main File -  Final Scraping

```python
name = []
wealth_history = []
personal_stats = []
x = file['internal_link'][1189:]
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36',"Content-Type": "application/x-www-form-urlencoded"} 
for i in x:
    print(i)
    url = i
    webpage = requests.get(url, headers=headers).text # fetching html webpage
    soup = BeautifulSoup(webpage, 'lxml')  # giving webpage to soup
    
    try:
        name.append(soup.find('h1','listuser-header__name').text) # name
    except:
        name.append(np.nan)
        
    try:
        wealth_history.append(soup.find('canvas').get('data-chart')) # wealth_history
    except:
        wealth_history.append(np.nan)
        
    try:
        personal_stats.append(soup.find('dl','listuser-block').text) # personal_stats
    except:
        personal_stats.append(np.nan)
        
print("Done")        
```

```python
internal_file_data = pd.DataFrame({
    "name":name,
    "Wealth_History":wealth_history,
    "personal_data":personal_stats
})
```

```python
internal_file_data.to_csv('internal_file_data.csv', encoding='utf-8')
```
